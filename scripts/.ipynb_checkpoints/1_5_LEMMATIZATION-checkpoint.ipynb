```python
### to do below some matching, we have to take care of appropriate encoding
from unicodedata import normalize
def normalize_string(string):
    try:
        return normalize("NFC", string)
    except: 
        return ""
PHI["string_normalized"] = PHI["clean_text_interpretive_word"].apply(normalize_string)
```

# Pythia data

```python
url = "SDAM_data/PHI/PHI_pythia/phi-plaintext.zip"
resp = conf[0].get(conf[1] + url)
```

```python
zipped = zipfile.ZipFile(io.BytesIO(resp.content))
```

```python
namelist = zipped.namelist()
namelist = [file for file in namelist if (".txt" in file) and (file[0] != "_")]
namelist[:10]
```

```python
phi_pythia_dict = {}
for filename in namelist:
    phi_number = filename.rpartition("/")[2].partition(".txt")[0]
    phi_string = normalize_string(zipped.read(filename).decode())
    phi_pythia_dict[phi_number] =  phi_string
```

```python
def get_pythia_text_by_number(phi_number):
    try: return phi_pythia_dict[str(phi_number)]
    except: return ""
```

```python
PHI["string_pythia"] = PHI["Text"].apply(get_pythia_text_by_number)  
```

```python
nlp = spacy.load("../models/spacy_grc_model_7")
path = "../models/spacy_grc_model_7/grc_lemmatization.py"
exec(open(path).read())
nlp.add_pipe(grc_doc_lemmatizer, "grc_doc_lemmatizer", after="tagger")
```

```python
def try_nlp(string):
    try: return nlp(string)
    except: return nlp("")
```

```python
%%time

## apply it on the normalized string
PHI_docs_series = PHI["string_normalized"].apply(try_nlp)
```

```python
def extraxt_lemmata(doc):
    lemmata = []
    for token in doc:
        if token.pos_ in ["NOUN", "PROPN", "ADJ", "VERB"]:
        lemmata.append(token.lemma_)
```

```python
lemmata_series = PHI_docs_series.apply(extraxt_lemmata)
```

```python
%%time
PHI["lemmata"] = lemmata_series
#PHI["lemmata"] = PHI["string_normalized"].apply(lambda x: lemmatize_string(x, all_lemmata=False, filter_by_postag=["n", "a", "v"], involve_unknown=True))
```
